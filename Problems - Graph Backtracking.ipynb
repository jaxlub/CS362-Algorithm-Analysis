# Word search

Implement your backtracking strategy for finding a word in a grid.

```python
def word_search(grid, word, path = None, word_orig = None):
  if path is None:
    path = list()
    word_orig = word

  if len(path) == len(word_orig):
    return path

  if len(path)==0:
    additions = list()
    for j in range(len(grid)):
      for i in range(len(grid[j])):
        if word[0] == grid[j][i]:
         additions.append((j,i))
  else:
    x,y = path[-1]
    word = word[1:]
    if len(word) == 0:
      return None
    next_let = word[0]
    additions = list()
    if x > 0 and grid[x-1][y] == next_let:
      additions.append((x-1,y))
    if x < len(grid)-1:
      if grid[x+1][y] == next_let:
        additions.append((x+1,y))
    if y < len(grid[x])-1:
      if grid[x][y+1] == next_let:
        additions.append((x, y+1))
    if y > 0 and grid[x][y-1] == next_let:
      additions.append((x,y-1))
  for x,y in additions:
    if (x,y) not in path:
      path.append((x,y))
      complete_path = word_search(grid, word, path, word_orig)

      if complete_path is not None:
        return complete_path
      else:
        path.pop()
```

```python
g = [
  ['B', 'A', 'A', 'C'],
  ['B', 'C', 'B', 'A'],
  ['C', 'C', 'B', 'C']
]
```

```python
print(word_search(g, "ABBCA"))
```
