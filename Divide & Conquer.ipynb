# Merge sort

This function implements our merge sort algorithm.

```python
# Merges two sorted regions of an array (from left to middle and from middle+1 to right).
def merge(array, left, middle, right):
  left_array = array[left : middle + 1]
  right_array = array[middle + 1 : right + 1]

  left_index = 0
  right_index = 0

  for i in range(left, right + 1):
    if right_index == len(right_array):
      array[i] = left_array[left_index]
      left_index += 1
    elif left_index == len(left_array):
      array[i] = right_array[right_index]
      right_index += 1
    elif left_array[left_index] < right_array[right_index]:
      array[i] = left_array[left_index]
      left_index += 1
    else:
      array[i] = right_array[right_index]
      right_index += 1
```

```python
# Sorts an array between the left and right indices (inclusive).
def recursive_merge_sort(array, left, right):
  if left < right:
    mid = (left + right) // 2
    recursive_merge_sort(array, left, mid)
    recursive_merge_sort(array, mid+1, right)
    merge(array,left,mid ,right)
```

```python
def merge_sort(array):
  left = 0
  right = len(array)-1
  recursive_merge_sort(array, left, right)
```

```python
ls = [7, 3, 2, 16, 24, 4, 11, 9]
merge_sort(ls)
print(ls)
```
