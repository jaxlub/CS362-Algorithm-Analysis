# Problem 1

Given an array of $n$ numbers, determine whether it contains a pair of numbers that add up to a target value.

A) Propose an efficient algorithm idea for this task.
Make an empty set and then loop through the values in the array. For each value calculate how far away from the target it is and then see if that value is present in our set. Afterwords if not present add it to the set. Return false if the loop is completed and no match has been found.

B) Analyze your algorithm.

This algorythm makes a set which is constant and then loops through the values which is N. It adds values to the set and looks up values too which is constant. This overall then has a O(n) but the ability to end early if the 2 first numbers sum to the target.

C) Implement your algorithm.

```python
def check_pair_sum(array, target):
  array_set=set()
  for value in array:
    looking_for = target - value
    if looking_for in array_set:
      return True
    array_set.add(value)
  return False
```

```python
print(check_pair_sum([1, 3, 2, 4], 6)) # 2 + 4 = 6
print(check_pair_sum([1, 2, 3, 4], 7)) # 3 + 4 = 7
```

```python
print(check_pair_sum([1, 2, 3, 4], 8))
print(check_pair_sum([1, 2, 3, 4], 9))
```

# Problem 2

Given a string of $n$ letters, determine whether it could be rearranged to form a palindrome.

A palindrome is the same sequence of letters both forwards and backwards.

A) Propose an efficient algorithm idea for this task.

Convert the array to a frequency counted dictionary (key = letter, value = count). Then if the array has an even number of letters make sure that the dictionary has all counts of even numbers (set of once forwards once back) and if odd make sure it has all even numbers and 1 odd (for the inflection point).

B) Analyze your algorithm.

Making a dict is constant. Looping through the array and adding to the dicitonary count is O(N). Then looping through the counts to make sure they are all the right numbers is O(N). There is no early outcut so this is Theta(N).

C) Implement your algorithm.

```python
def check_rearranged_palindrome(s):
  count = dict()
  for letter in s:
    if letter in count:
      count[letter] += 1
    else:
      count[letter]=1

  if (len(s) % 2) == 1:
    odds = 0
    for (key, value) in count.items():
      if count[key] == 1:
        odds += 1
    if odds != 1:
      return False
  else:
    odds = 0
    for key, value in count.items():
      if count[key] == 1:
        odds += 1
    if odds != 0:
      return False
  return True
```

```python
print(check_rearranged_palindrome("aabb")) # abba
print(check_rearranged_palindrome("aab")) # aba
```

```python
print(check_rearranged_palindrome("ab"))
print(check_rearranged_palindrome("aaajbb"))
```
