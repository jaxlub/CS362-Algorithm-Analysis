# Problem 1

Use a key to sort the numbers without regard to [sign](https://docs.python.org/3/library/functions.html#abs): `[0, -1, 1, -2, 2, -3, 3]`

```python
ls = [-3, -2, -1, 0, 1, 2, 3]


print(sorted(ls, key=lambda item: abs(item)))
```

Use a key to sort the strings without regard to [case](https://docs.python.org/3/library/stdtypes.html?highlight=lower#str.lower): `['brown', 'Fox', 'quick', 'The']`

```python
ls = ['The', 'quick', 'brown', 'Fox']


print(sorted(ls, key=lambda item: item.lower()))
```

Use a key to put evens before odds: `[2, 16, 24, 4, 7, 3, 11, 9]`

```python
ls = [7, 3, 2, 16, 24, 4, 11, 9]


print(sorted(ls, key=lambda item: item%2))
```

Use a key to find the string with the largest [integer value](https://docs.python.org/3/library/functions.html#int): `24`

```python
ls = ["7", "3", "2", "16", "24", "4", "11", "9"]


print(max(ls, key=lambda item: int(item)))
```

Use a key to print the shortest (start, end) time interval: `(2, 3)`

```python
ls = [(2, 3), (2, 4), (1, 5), (1, 6)]

print(min(ls, key= lambda item: item[1]-item[0]))
```

# Problem 2



A) Complete the function below to return the index of the largest value in the array, without using any of our new tricks (yet).

```python
def max_index(array):
  max_index = 0
  for i in range(len(array)):
    if array[i] > array[max_index]:
      max_index = i
  return max_index
```

```python
ls = [7, 3, 2, 16, 24, 4, 11, 9]
print(max_index(ls))
```
