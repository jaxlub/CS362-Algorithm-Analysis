# Problem 1

Suppose we want to look at shortest paths from one vertex to all other reachable vertices.

Complete the `all_shortest_paths` generator function to yield these paths one at a time.

Avoid traversing the graph more than once.

```python
def all_shortest_paths(graph, start):
    from collections import deque
    queue = deque([start])
    parent = {start: None}

    while len(queue) > 0:
      vertex = queue.popleft()
      for adjacent in graph[vertex]:
        if adjacent not in parent:
          parent[adjacent] = vertex
          queue.append(adjacent)
          #reconstruct path for whichever we're on
          v = adjacent
          path = [adjacent]
          while v != start:
            v = parent[v]
            path.append(v)
          path.reverse()
          yield path
```

```python

for path in all_shortest_paths(g, 1):
  print(path)


```

# Problem 2

Suppose we want to look at all the components of a graph.

Complete the `all_components` generator function to yield them one at a time.

Avoid traversing any component more than once.

```python
# Our function for collecting a single component will be useful here.
def component(graph, vertex, discovered=None):
  if discovered is None:
    discovered = {vertex}
  else:
    discovered.add(vertex)

  for neighbor in graph[vertex]:
    if neighbor not in discovered:
      component(graph, neighbor, discovered)

  return discovered
```

```python
def all_components(graph):
  discovered = set()

  for vertex in graph:
    if vertex not in discovered:
      peice = component(graph, vertex)
      for value in peice:
        discovered.add(value)
      yield peice

```

```python

for comp in all_components(g):
  print(comp)
```
