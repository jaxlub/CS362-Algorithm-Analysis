# Problem 1

Suppose a computer network is described to you as a list of `(x, y, t)` tuples.

In each tuple, `x` and `y` are computers and `t` is the time it takes for a message to travel along a direct link between `x` and `y`.

One of the computers is about to broadcast a message along all of its direct links.

The first time any other computer receives a broadcast, it will propagate the message along all of its own direct links.

How much time will it take until every computer in the network has received the broadcast?

A) Work out an answer yourself for the example below and explain your thinking.

- Network: `[(A, B, 1), (A, C, 4), (B, C, 1), (B, D, 2), (C, D, 3)]`

- Broadcast origin: `A`

3 units of time - A to B is 1 and then from there it is one T to C and 2 Ts to D and thus all are covered.

B) Give an algorithm idea for solving this problem in general.

First make a graph, where each vertex goes to its corresponding vertex with the edge as the time weight and vice versa (so A goes to B in 1 and B goes to A in 1). Then once construcuted use dijsktras algorithm to add up the lightest weights to hit every single vertex and return the value of the last thing in the queue which weight is the total weight it takes to get there (which is the lightest path to all points). This a little different from the one we wrote in class that doesnt stop at the "end" vertex and does them all.

C) Give (and justify) an upper bound for your algorithm.

We loop through every value first to construct the graph which is time of E. Then we run dijskstras algorithm which is O((E+V)logv) which dominates the time to construct the graph so overall it is O((E+V)logv).

```python
def broadcast_time(origin, network):
  from heapdict import heapdict
  from math import inf
  graph = {}
  for values in network:
    v1, v2, weight = values

    if v2 not in graph:
        graph[v2] = {}

    if v1 not in graph[v2]:
        graph[v2][v1] = {}

    graph[v2][v1] = weight

    if v1 not in graph:
        graph[v1] = {}

    if v2 not in graph[v1]:
        graph[v1][v2] = {}

    graph[v1][v2] = weight
  print(graph)

  # set up data structures
  priority = heapdict({v: inf for v in graph})
  parent = {origin: None} # Start has no parent as it is the root
  priority[origin] = 0
  while len(priority) > 0:
    vertex, vertex_priority = priority.popitem()
    for adjacent in graph[vertex]:
      if adjacent in priority:
        new_priority = vertex_priority + graph[vertex][adjacent] # coming from current vertex and add pathwieght to next

        if new_priority < priority[adjacent]:
          priority[adjacent] = new_priority
          parent[adjacent] = vertex
  return vertex_priority
```

```python
print(broadcast_time('A', [('A', 'B', 1), ('A', 'C', 4), ('B', 'C', 1), ('B', 'D', 2), ('C', 'D', 3)]))
```
