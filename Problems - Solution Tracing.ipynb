C) Describe how you will extract the optimal EMP indices once the arrays are complete.

Once the C array is constructed we will go back through the array backwards as the last value is optimal and each time we hit a 1 we will take note of this wave and then skip back k waves to account for the recharge time.

# Problem 2

Implement your dynamic programming algorithm with solution tracing.

```python
def optimal_EMP_indices(a, k):
  n = len(a) # Number of waves
  v = [0 for i in range(n)]
  c = [0 for i in range(n)]

  v[0] = a[0]
  c[0] = 1

  for i in range(1, n):
    # price of blowing up this wave
    if i < k:
      v[i], c[i] = max((a[i],1), (v[i-1],0))
    else:
      v[i], c[i] = max((a[i]+v[i-k], 1), (v[i-1], 0))


  choices = []
  i = len(c)-1
  while i > -1:
    if c[i] == 1:
      choices.append(i)
      i = i-k
    else:
      i = i-1
  choices.reverse()
  return choices

```

```python
print(optimal_EMP_indices([1,4,3,6,5,2], 2))
print(optimal_EMP_indices([6,3,5,2,4,1], 3))
```
