# Stair climbing

This function returns a sequence of step sizes for minimizing noise as we go up the stairs.

```python
def optimal_step_sizes(a):
  n = len(a) # Number of stairs
  a.append(0) # Top landing

  v = [None for i in range(n+1)]
  c = [None for i in range(n+1)]

  v[0], c[0] = a[0], 1
  v[1], c[1] = a[1], 2

  for i in range(2, n+1):
    v[i], c[i] = min((v[i-2] + a[i], 2), (v[i-1] + a[i],1))

  choices = []
  i = n
  while i >= 0:
    choices.append(c[i])
    i = i - c[i]
  choices.reverse()
  return choices
```

```python
print(optimal_step_sizes([1, 1, 1, 1, 1]))
print(optimal_step_sizes([1, 9, 1, 9, 1]))
print(optimal_step_sizes([9, 1, 1, 9, 1]))
```

# Change making

This function returns a minimal list of coins needed to make change.

```python
def optimal_change(n, coins):
  v = [None for i in range(n+1)]
  c = [None for i in range(n+1)]

  v[0], c[0] = 0, None

  for i in range(1, n+1):
    v[i], c[i] = min((1 + v[i - coin], coin) for coin in coins if coin <= i)

  choices = []
  i = len(v)-1
  while i > 0:
    choices.append(c[i])
    i = i - c[i]
  choices.reverse()
  return choices
  #return v[n]
```

```python
print(optimal_change(6, {25, 10, 5, 1}))
print(optimal_change(12, {25, 10, 5, 1}))
print(optimal_change(31, {25, 10, 5, 1}))
```
