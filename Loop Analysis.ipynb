# Selection sort

This function implements the selection sort algorithm.

```python
def selection_sort(array):
  for i in range(len(array)):
    smallest_index = i
    for y in range(i+1, len(array)):
      if array[y] < array[smallest_index]:
        smallest_index = y
    array[smallest_index], array[i]= array[i], array[smallest_index]
```

```python
ls = [7, 3, 2, 16, 24, 4, 11, 9]
selection_sort(ls)
print(ls)
```

# Insertion sort

This function implements the insertion sort algorithm.

```python
def insertion_sort(array):
  for i in range(1, len(array)):
    while array[i] < array[i-1] and i > 0:
      array[i], array[i-1] = array[i-1], array[i]
      i -= 1
```

```python
ls = [7, 3, 2, 16, 24, 4, 11, 9]
insertion_sort(ls)
print(ls)
```
