# Fibonacci numbers

This function computes the nth Fibonacci number using the recursive definition.

```python
def f(n):
  if n <= 1:
    return n
  else:
    return f(n-1) + f(n-2)
```

```python
print(f(10))
```

This function computes the nth Fibonacci number using the dynamic programming algorithm.

```python
def fib(n):
  answer = list()
  answer.append(0)
  answer.append(1)

  for i in range(n-1):
    i = 0
    i = answer[-1] + answer[-2]
    answer.append(i)
  return(answer[len(answer)-1])

```

```python
print(fib(100))
```

# Stair climbing

This function returns the minimum amount of noise we can make going up the stairs.

```python
def min_stair_noise(a):
  n = len(a)
  a.append(0)

  v = [None for i in range(n+1)]
  v[0] = a[0]
  v[1] = a[1]

  for i in range(2,n+1):
    v[i] = min(v[i-2]+a[i], v[i-1]+ a[i])
  return v[n]


```

```python
print(min_stair_noise([1, 1, 1, 1, 1]))
print(min_stair_noise([1, 9, 1, 9, 1]))
print(min_stair_noise([9, 1, 1, 9, 1]))
```

# Change making

This function returns the minimum number of coins needed to make change.

```python
def min_coin_number(n, coins):
  v = [None for i in range(n+1)]

  v[0] = 0
  for i in range(1,n+1):
    v[i] = min([1+v[i-coin] for coin in coins if coin<=i])
  return v[n]
```

```python
print(min_coin_number(6, {25, 10, 5, 1}))
print(min_coin_number(12, {25, 10, 5, 1}))
print(min_coin_number(31, {25, 10, 5, 1}))
```

```python
print(min_coin_number(6, {25, 10, 6, 1}))
print(min_coin_number(12, {25, 10, 6, 1}))
print(min_coin_number(31, {25, 10, 6, 1}))
```
