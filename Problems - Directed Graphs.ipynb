B) Give an algorithm idea for solving this problem in general.

Convert the series of tuples (pre-req, class) into a dictionary where the class is the key and the pre-req is the value making a directed graph. Then using the algorithm we implmented for the directed graph: run DFS on the goal and then all of its adjacents until theres nothing left and as we do this add them to a list as we back out to return at the end.

# Problem 2

Complete the function below to return a course sequence.

```python
from IPython.core.interactiveshell import dis
def course_sequence(goal, prereqs):
  # Make Directed Graph
  graph = dict()
  for prereq in prereqs:
    pre_req, class_ = prereq
    if class_ not in graph:
      graph[class_] = []
    graph[class_].append(pre_req)
  discovered = set()
  result = []


  def dfs(v): # single DFS from vertex v
    discovered.add(v)
    if v in graph:
      for adjacent in graph[v]:
        if adjacent not in discovered:
          dfs(adjacent)
    result.append(v)




    #Run dfs on 362
  dfs(goal)


  return result

#{'CS140': {'CS219'}, 'CS219': {'CS256', 'CS220'}, 'CS256': {'CS362', 'CS374'}, 'MA280': {'CS362', 'CS380'}}
#{'CS140': set(), 'CS219': {'CS140'}, 'CS256': {'CS219'}, 'MA280': set(), 'CS220': {'CS219'}, 'CS362': {'CS256', 'MA280'}, 'CS374': {'CS256'}, 'CS380': {'MA280'}}

```

```python
print(course_sequence("CS362", [("CS140", "CS219"),("CS219", "CS220"), ("CS219", "CS256"), ("CS256", "CS362"), ("CS256", "CS374"), ("MA280", "CS362"), ("MA280", "CS380")]))
```
