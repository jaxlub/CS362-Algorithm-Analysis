# Target

This function determines whether an array contains a target value.

```python
def check_target(array, target):
  for value in array:
    if value == target:
      return True
  return False
```

```python
print(check_target([1, 2, 3], 1))
print(check_target([1, 2, 3], 3))
print(check_target([1, 2, 3], 5))
```

# Largest

This function finds the largest value in an array.

```python
def find_largest(array):
  largest = array[0]
  for i in range(1, len(array)):
    if array[i] > largest:
      largest = array[i]
  return largest
```

```python
print(find_largest([1, 2, 3]))
print(find_largest([3, 2, 1]))
print(find_largest([1, 3, 2]))
```

# Identical

This function checks whether two arrays are identical.

```python
def check_identical(A, B):
  if len(A) != len(B):
    return False
  for i in range(len(A)):
    if A[i] != B[i]:
      return False
  return True
```

```python
print(check_identical([1, 2, 3], [1, 2, 3]))
print(check_identical([1, 2, 3], [1, 3, 2]))
print(check_identical([1, 2, 3], [1, 2, 3, 4]))
```

# Matches

This function counts the ways to choose matching values from two arrays.

```python
def count_matches(A, B):
  counter = 0
  for valueA in A:
    for valueB in B:
      if valueA == valueB:
        counter = counter + 1
  return counter
```

```python
print(count_matches([1, 2, 3], [1, 2, 3]))
print(count_matches([1, 1, 2], [1, 2, 2]))
print(count_matches([1, 1, 1], [1, 1, 1]))
```
