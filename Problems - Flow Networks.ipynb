```python
from re import X
from pulp import *
lp = LpProblem("flow", LpMaximize)

SA = LpVariable("SA",0,2)
SB = LpVariable("SB",0,2)
SD = LpVariable("SD",0,2)
SC = LpVariable("SC",0,2)

AZ = LpVariable("AZ",0,1)
AX = LpVariable("AX",0,1)
AY = LpVariable("AY",0,1)
BW = LpVariable("BW",0,1)
BP = LpVariable("BP",0,1)
BV = LpVariable("BV",0,1)
CZ = LpVariable("CZ",0,1)
CX = LpVariable("CX",0,1)
CY = LpVariable("CY",0,1)
CW = LpVariable("CW",0,1)
CP = LpVariable("CP",0,1)
CV = LpVariable("CV",0,1)
DW = LpVariable("DW",0,1)
DP = LpVariable("DP",0,1)
DV = LpVariable("DV",0,1)
DL = LpVariable("DL",0,1)
DH = LpVariable("DH",0,1)




ZF = LpVariable("ZF",0,1)
XF = LpVariable("XF",0,1)
YF = LpVariable("YF",0,1)
WF = LpVariable("WF",0,1)
PF = LpVariable("PF",0,1)
VF = LpVariable("VF",0,1)
LF = LpVariable("LF",0,1)
HF = LpVariable("HF",0,1)


# add coservation constraints
lp += (SA == AZ + AX + AY)
lp += (SB == BW + BP + BV)
lp += (SC == CZ + CX + CY + CW + CP + CV)
lp += (SD == DW + DP + DV + DL + DH)
lp += (AZ + CZ == ZF)
lp += (CX + AX == XF)
lp += (CY == YF)
lp += ( DW + CW + BW == WF)
lp += (DP+ CP + BP == PF)
lp += (DV + CV + BV == VF)
lp += (DL == LF)
lp += (DH == HF)





# add f paths
# Defining the objective function
lp += (ZF + XF + YF + WF + PF + VF + LF + HF)


```

```python
# into many cells
lp.solve()


```

```python

# Checking the flows along the edges
for edge in lp.variables():
  print(edge, "=", value(edge))
```
