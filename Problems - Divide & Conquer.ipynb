```python
def f(array):
  n = len(array)
  f(array[0 : n//3])
  f(array[2*n//3 : n])
```

# Problem 2

Analyze this function using the master theorem.

```python
def f(array):

  def g(array, left, right):
    middle = (left + right) // 2
    g(array, left, middle)
    g(array, left, middle)
    g(array, middle + 1, right)
    g(array, middle + 1, right)

  g(array, 0, len(array) - 1)
```

A) Write a recurrence relation to express the growth rate.
T(N)=4T(n/2)+1

B) Show how to solve the recurrence relation to compute the upper bound.

$aT(n/b)+n^d$ \
a = 4 \
b = 2 \
d = 0 (since n^0 is 1)\

log base b of a AKA log base 2 of 4 is 2 which is more then d of 0 so this has a time complexity of $O(n^(log B of A))$ or $O(n^2)$.
