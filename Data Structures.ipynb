# Problem 1

The functions below demonstrate a few options for implementing LIFO stacks and FIFO queues.

Use timing experiments and/or internet/AI research to compare them and select the better options.

```python
# We could use a list as a LIFO stack.
def list_stack(n):
  stack = []

  # Push values on the end
  for i in range(n):
    stack.append(i)

  # Pop them off the end
  while len(stack) > 0:
    stack.pop()
```

```python
# We could use a deque as a LIFO stack.
from collections import deque

def deque_stack(n):
  stack = deque([])

  # Push values on the end
  for i in range(n):
    stack.append(i)

  # Pop them off the end
  while len(stack) > 0:
    stack.pop()
```

A) Which stack implementation is preferable, and why?

Deque is prefereable becuase while addding and removing from both have the same run time, a deque enver has to reallocate to a new array like if a list becomes full, which prevents it having to do extra linear tasks.


```python
# We could use a list as a FIFO queue.
def list_queue(n):
  queue = []

  # Push values on the end
  for i in range(n):
    queue.append(i)

  # Pop them off the front
  while len(queue) > 0:
    queue.pop(0)
```

```python
# We could use a deque as a FIFO queue.
from collections import deque

def deque_queue(n):
  queue = deque([])

  # Push values on the end
  for i in range(n):
    queue.append(i)

  # Pop them off the front
  while len(queue) > 0:
    queue.popleft()
```

B) Which queue implementation is preferable, and why?

The list in this case must shift everything down to the index 0 on a pop which requires shfiting the whole array (O(n)) where as the deque does not and can still keep popping off the front for a time complexity of O(1).

A) Complete the function below to create and return a dictionary.

Each key in the dictionary is a letter that appears in the string argument.

The value associated with each key is the number of times it appears in the string.

```python
def letter_count(s):
  count = dict()
  for i in range(len(s)):
    if s[i] in count:
      count[s[i]]+=1
    else:
      count[s[i]] =1
  return count
```

```python
print(letter_count("aaabbc"))
```

B) Add a key below to put the letters in order of their frequency: `['c', 'b', 'a']`

```python
count = letter_count("aaabbc")
print(sorted(count.keys(), key = lambda item:count[item]))
```

C) Add a comprehension below to print a set of all the letters that appear more than once: `{'a', 'b'}`

```python
count = letter_count("aaabbc")

multiple = {value for value in count if count[value] > 1 }
print(multiple)
```
