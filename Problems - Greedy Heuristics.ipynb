# Problem 1

Suppose we want to select integers from an array to produce a sum as close as possible to a target without going over.

For example, given `[4, 8, 2, 9]` and a target sum of 7, the best we can do is select `[4, 2]` which add up to 6.

This example suggests a greedy heuristic: repeatedly select the largest integer that doesn't make our sum exceed the target.

A) Give a counterexample to show that this greedy heuristic isn't correct.

If we had [10,6,5] and were trying to make 11 the HG would suggest 10 but a better solution would be 5 and 6.


B) Implement this greedy heuristic.

```python
def greedy_closest_sum(array, target):
  answer = list()
  for number in sorted(array, reverse=True):
    if number <= target:
      answer.append(number)
      target = target - number
  return answer
```

```python
print(greedy_closest_sum([4, 8, 2, 9], 7)) # Should return [4, 2]
print(greedy_closest_sum([4, 7, 2, 9], 7)) # Should return [7]
print(greedy_closest_sum([4], 8))
```

# Problem 2

Suppose we want to select as many integers as possible from an array without going over a sum limit.

For example, given `[1, 5, 2, 3]` and a sum limit of 7, the best we can do is select `[1, 2, 3]` which add up to 6.

This example suggests a greedy heuristic: repeatedly select the smallest integer that doesn't make our sum exceed the limit.

A) Give a convincing argument that this greedy heuristic does always work.
Repeatdly select the smallest values in the list without repeating for as long as you dont exceed the limit. This maximizes the numbers used by choosing the smallest one each time and thus allows for maximum "room" left before hitting the limit.

B) Implement this greedy heuristic.

```python
def greedy_most_integers(array, limit):
  answer = list()
  for number in sorted(array):
    if number <= limit:
      answer.append(number)
      limit = limit - number
  return answer
```

```python
print(greedy_most_integers([1, 5, 2, 3], 7)) # Should return [1, 2, 3]
print(greedy_most_integers([4, 5, 2, 3], 7)) # Should return [2, 3]
```
