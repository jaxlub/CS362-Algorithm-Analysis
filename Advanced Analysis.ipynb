# Problem 1

This function attempts to implement binary search recursively.

It works, as you can see, but it isn't as efficent as binary search ought to be.

Note: the slice operation `a[i:j]` copies a subarray of `a` from index `i` (inclusive) to index `j` (exclusive).

```python
def slow_recursive_binary_search(array, target):
  if len(array) == 0:
    return False

  mid = len(array) // 2

  if array[mid] > target:
    return slow_recursive_binary_search(array[mid+1:len(array)], target)
  elif array[mid] < target:
    return slow_recursive_binary_search(array[0:mid], target)
  else:
    return True
```

```python
ls = [1, 3, 5, 7]
for t in range(9):
  print(t, slow_recursive_binary_search(ls, t))
```

Give (and justify) an upper bound for the `slow_recursive_binary_search` function.

The upperbound case for this algorithm is when looking for a number not present in the array and thus it recurses the max amount of times to get to a subset of 0 which is splitting it in half which can be done log base 2 of n iterations which is O(log(n)). The slicing is only done once but is O(n) resulting in O(N+logN) which is really O(N) overall.

# Problem 2

Implement a recursive binary search that *is* as efficient as it ought to be.

```python
def binary_search(array, target):
  if len(array) == 0:
    return False
  left = 0
  right = len(array)-1
  return binary_search_helper(array, target, left, right)

def binary_search_helper(array, target, left_bound, right_bound):
  mid_index = (left_bound + right_bound) // 2

  if left_bound > right_bound:
    return False
  #print(array, mid_index, left_bound, right_bound)
  if array[mid_index] < target:
    return binary_search_helper(array, target, mid_index+1, right_bound)
  elif array[mid_index] > target:
    return binary_search_helper(array, target, left_bound, mid_index-1)
  else:
    return True
```

```python
ls = [1, 3, 5, 7]
for t in range(9):
  print(t, binary_search(ls, t))
```
