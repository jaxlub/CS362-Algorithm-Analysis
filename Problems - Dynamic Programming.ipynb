# Problem 1

Robots are coming to attack our classroom! We have detected them ahead of time, so we know there will be one wave of robots each minute for `n` minutes, and we know how many robots will be in each wave. We have stored the sequence of wave sizes in an array called `a`.

Our primary weapon is an electro-magnetic pulse (EMP). Activating it will destroy all the robots in one wave. But we can't just activate it for every wave, because it requires at least `k` minutes between uses to recharge.

We can't allow robots to accumulate, so any wave not destroyed with the EMP must be deactivated through perilous hand-to-hand combat before the next wave arrives. Which leads to our question: **What is the maximum number of robots we can destroy with the EMP?**

Here are a few examples:
- Given `a = [1,4,3,6,5,2]` and `k = 2`, we can destroy at most `4+6+2 = 12` robots.
- Given `a = [6,3,5,2,4,1]` and `k = 3`, we can destroy at most `6+4 = 10` robots.

And here is some notation:
- Let $v[i]$ be the value of the solution for subproblem $i$ where $0 \leq i \leq n$.
- So $v[i]$ is the maximum number of robots we can destroy throughout the first $i$ waves.

# Problem 2

Implement your dynamic programming algorithm.

```python
def max_destroyable_robots(a, k):

  n = len(a) # Number of waves
  v = [0 for i in range(n)]

  v[0] = a[0]

  for i in range(1, n):
    # price of blowing up this wave
    if i < k:
      v[i] = max(a[i], v[i-1])
    else:
      v[i] = max(a[i]+v[i-k], v[i-1])

  return(v)


```

```python
print(max_destroyable_robots([1,4,3,6,5,2], 2))
print(max_destroyable_robots([6,3,5,2,4,1], 3))
```
