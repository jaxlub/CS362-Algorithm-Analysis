# Flood fill

This function fills a clicked region of an image with a color.

```python
def flood_fill(image, click, new_color):
  (r,c) = click
  if image[r][c] != new_color:
    old_color = image[r][c]
    image[r][c] = new_color


    if r > 0 and image[r-1][c] == old_color: # up
      flood_fill(image, (r-1, c), new_color)

    if r < len(image) -1 and image[r+1][c] == old_color: # down
      flood_fill(image, (r+1, c), new_color)

    if c < len(image[r]) -1 and image[r][c+1] == old_color: # right
      flood_fill(image, (r, c+1), new_color)

    if c > 0 and image[r][c-1] == old_color: # left
      flood_fill(image, (r, c-1), new_color)






```

```python
m = [
  [1,1,1],
  [1,1,0],
  [1,0,1]
]

flood_fill(m, (1, 1), 2)
for row in m:
  print(row)
```

```python
m = [
  [1,1,1],
  [1,1,0],
  [1,0,1]
]

flood_fill(m, (1, 1), 1)
for row in m:
  print(row)
```

```python
m = [
  [1,1,1],
  [1,1,0],
  [1,0,1]
]

flood_fill(m, (1, 1), 0)
for row in m:
  print(row)
```

# Word ladder

This function finds the shortest word ladder between two words through a set of valid words.

```python
def word_ladder(start, end, valid):
  if start not in valid or end not in valid:
    return None
  if start == end:
    return [start]
  from collections import deque
  if start == end:
    return [start]
  queue = deque([start])
  parent = {start: None}

  while len(queue) > 0:
    word = queue.popleft()

    for i in range(len(word)):
      for letter in "abcdefgihjklmnopqrstuvwxyz":
        adjacent = word[:i]+letter + word[i+1:]
        if adjacent in valid and adjacent not in parent:
          parent[adjacent] = word
          queue.append(adjacent)
          if adjacent == end:
            # reconstruct path
            v = end
            path = [end]
            while v != start:
              v = parent[v]
              path.append(v)
            path.reverse
            return path
```

```python
from requests import get
url = "https://myslu.stlawu.edu/~ltorrey/files/362/words.txt"
words = set(get(url).text.split()) # 370k English words
```

```python
print(word_ladder("cold", "warm", words))
```

```python
print(word_ladder("winter", "summer", words))
```
