# Maximum flow

This code solves a linear program to maximize flow in the graph.

```python
# Installing the module
!pip install pulp
```

```python
# Initializing the linear program
from pulp import *
lp = LpProblem("flow", LpMaximize)
```

```python
# Defining a variable for each edge
SA = LpVariable("SA", 0, 7)
SD = LpVariable("SD",0,4)
DA = LpVariable("DA", 0,3)
DC = LpVariable("DC", 0, 2)
AC = LpVariable("AC", 0, 3)
AB = LpVariable("AB", 0, 5)
CB = LpVariable("BC", 0 , 3)
CT = LpVariable("CT", 0 , 5)
BT = LpVariable("BT", 0, 8)
```

```python
# Defining the conservation constraints
lp += (SA + DA == AB + AC)
lp += (AB + CB == BT)
lp += (AC + DC == CB + CT)
lp += (SD == DA + DC)
```

```python
# Defining the objective function
lp += (BT + CT)
```

```python
# Solving the problem
lp.solve()
```

```python
# Checking the total flow
value(lp.objective)
```

```python
# Checking the flows along the edges
for edge in lp.variables():
  print(edge, "=", value(edge))
```
