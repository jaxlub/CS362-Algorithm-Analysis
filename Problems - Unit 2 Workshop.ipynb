# Problem 1

Suppose you are given two water jugs, which hold exactly `a` and `b` gallons, and asked to measure exactly `c` gallons of water.

The steps you can take are:
- Fill either jug from a faucet (until it's completely full).
- Dump either jug down the drain (until it's completely empty).
- Transfer water from one jug to the other (until one is empty or the other is full, whichever happens first).

You want to complete the task as quickly as possible.

What sequence of steps should you take?

A) Work out an answer yourself for the example below and explain your thinking.
- a = 5
- b = 3
- c = 4

Fill the 5 and dump it in into the 3. Now dump the 3 out (and the 5 has 2). Then move the 2 to the 3 and 5 is empty. Now Fill the 5 again and top off the 3 leaving 4 in the 5 bucket.

B) Give an algorithm idea for solving this problem in general.

Make a graph where the initial point is 0,0 and corresponds to each of the buckets. Then make another point for filling bucket 1 (x,0), filling bucket 2 (y,0), moving bucket 1 to 2 and 2 to 1 (x-y, y) and (x, y-x) (depends on which is bigger), emptying each bucket (x,0), (0,y). Then we have a very complex graph and can traverse with a BFS (to get the shortest possible path) where x or y = c.

# Problem 2

Complete the function below to print a sequence of instructions.

```python
def measure_water(a, b, c):
  from collections import deque
  if a == c or b == c:
    return "done"
  queue = deque()
  queue.append((0,0))
  parent = {(a,b):None}

  while len(queue) > 0:
    #print("next")
    adjacents = []
    Acurrent, Bcurrent = queue.popleft()

    if Acurrent != a:
      # fill A
      adjacents.append((a, Bcurrent))
      #print("fill a", a, Bcurrent, Acurrent, Bcurrent)

    if Bcurrent != b:
      # fill B
      adjacents.append((Acurrent, b))
      #print("fill b", Acurrent, b, Acurrent, Bcurrent)

    if Acurrent != 0:
      # empty A
      adjacents.append((0,Bcurrent))
      #print("empty a", 0, Bcurrent, Acurrent, Bcurrent)

    if Bcurrent != 0:
      # empty B
      adjacents.append((Acurrent,0))
      #print("empty b", Acurrent, 0, Acurrent, Bcurrent)

    if Acurrent != 0 and Bcurrent != b:
      # A can be poured into B
      Bcurrenttemp = Bcurrent + Acurrent
      if Bcurrenttemp > b:
        Acurrenttemp = Bcurrenttemp - b
        Bcurrenttemp = b
      else:
        Acurrenttemp = 0
      adjacents.append((Acurrenttemp, Bcurrenttemp))
      #print("a to b", Acurrenttemp, Bcurrenttemp, Acurrent, Bcurrent)



    if Acurrent != a and Bcurrent != 0:
      # B can be poured into A
      Acurrenttemp = Acurrent + Bcurrent
      if Acurrenttemp > a:
        Bcurrenttemp = Acurrenttemp - a
        Acurrenttemp = a
      else:
        Bcurrenttemp = 0
      adjacents.append((Acurrenttemp, Bcurrenttemp))
      #print("b to a", Acurrenttemp, Bcurrenttemp, Acurrent, Bcurrent)

    for adjacent_val in adjacents:
      acurr,bcurr = adjacent_val
      if adjacent_val not in parent:
        parent[adjacent_val] = (Acurrent,Bcurrent)
        queue.append(adjacent_val)
        if acurr == c or bcurr == c:
          # reconstruct path
          v = adjacent_val
          path = [adjacent_val]
          while v != (0,0):
            v = parent[v]
            path.append(v)
          path.reverse()
          return path
```

```python
measure_water(5, 3, 4)
```
