# Problem 1

Implement our dynamic programming algorithm to compute the edit distance between two strings.

```python
def edit_distance(a, b):
  def diff(c,d):
    if c == d:
      return 0
    else:
      return 1
  v = [[0] * (len(a)+1) for x in range(len(b)+1)]
  for i in range(0,len(a)+1):
    v[0][i] = i
  for i in range(0, len(b)+1):
    v[i][0] = i
  for i in range(1,len(v)):
    for j in range(1,len(v[i])):
      v[i][j] = min(((v[i-1][j]) +1),
                    ((v[i][j-1]) +1),
                    ((v[i-1][j-1]) + diff(a[i-1],b[j-1])))
  fin_row = v[len(v)-1]

  return fin_row[len(fin_row)-1]
```

```python
print(edit_distance('ACGT', 'TACG'))
```

```python
print(edit_distance('ACGTACGT', 'TACGAACG'))
```

Add solution tracing to our algorithm to compute the optimal alignment of two strings.

```python
def optimal_alignment(a, b):
  def diff(c,d):
    if c == d:
      return 0
    else:
      return 1
  v = [[0] * (len(a)+1) for x in range(len(b)+1)]
  c = [[0] * (len(a)+1) for x in range(len(b)+1)]
  for i in range(0,len(a)+1):
    v[0][i] = i
    c[0][i] = "-"+b[i-1]
  for i in range(0, len(b)+1):
    v[i][0] = i
    c[i][0] = a[i-1]+"-"
  for i in range(1,len(v)):
    for j in range(1,len(v[i])):
      v[i][j],c[i][j] = min((((v[i-1][j]) +1), a[i-1]+"-"),
                            (((v[i][j-1]) +1), "-"+b[j-1]),
                            (((v[i-1][j-1]) + diff(a[i-1],b[j-1]), a[i-1]+b[j-1])))

  for rows in v:
    print(rows)
  for rows in c:
    print(rows)

  # get last move in C
  last_cordx = len(c)-1
  #print(last_cordx)
  last_cordy = len(c[last_cordx])-1
  #print(last_cordy)

  choices = []
  choices.append(c[last_cordx][last_cordy])

  # max number of steps we can have is all the way up and then all the way across aka len(a)-1+len(b-1)
  while last_cordx > 0 and last_cordy > 0:
    if c[last_cordx][last_cordy] == 0:
      choices.append(c[last_cordx][last_cordy])
      break
    if c[last_cordx][last_cordy].endswith("-"):
        last_cordx = last_cordx-1
        choices.append(c[last_cordx][last_cordy])
    if c[last_cordx][last_cordy].startswith("-"):
        last_cordy = last_cordy-1
        choices.append(c[last_cordx][last_cordy])
    else:
        last_cordy = last_cordy-1
        last_cordx = last_cordx-1
        choices.append(c[last_cordx][last_cordy])
  choices.reverse()
  return choices
  return
```
