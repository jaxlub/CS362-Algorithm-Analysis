# Overlap

Given two arrays of length $n$, count how many values appear in both arrays.

```python
def count_overlap(A, B):
  count = 0
  #Aset = set()
  #Bset = set()
  #for a in A:
  #   Aset.add(a)
  #for b in B:
  #   Bset.add(b)

  #setA = {a for a in A}
  A = set(A)
  B = set(B)

  for value in A:
    if value in B:
      count += 1
  return count
```

```python
print(count_overlap([1, 2, 3, 4], [5, 6, 7, 8]))
print(count_overlap([1, 2, 3, 4], [4, 5, 6, 7]))
print(count_overlap([1, 2, 3, 4], [3, 4, 3, 4]))
print(count_overlap([1, 2, 3, 1], [1, 2, 3, 4]))
```

# Duplicates

Given an array of length $n$, determine whether it contains any duplicate values.

```python
def check_duplicates(array):
  #return len(array) != len(set(array)) Theta(N)
  #for omega(1) and O(n)
  seen = set()
  for value in array:
    if value in seen:
      return True
    else:
      seen.add(value)
  return False
```

```python
print(check_duplicates([1, 1, 2, 3]))
print(check_duplicates([1, 2, 1, 3]))
print(check_duplicates([1, 2, 3, 1]))
print(check_duplicates([2, 1, 1, 3]))
print(check_duplicates([2, 1, 3, 1]))
print(check_duplicates([2, 3, 1, 1]))
```

```python
print(check_duplicates([]))
print(check_duplicates([1]))
print(check_duplicates([1, 2]))
print(check_duplicates([1, 2, 3]))
```

# Unique

Given an array of length $n$, find the first index of any unique value.

If there is no unique value, return -1.

```python
def find_first_unique_index(array):
  freq_dict = dict()
  for value in array:
    if value in freq_dict:
      freq_dict[value] += 1
    else:
      freq_dict[value] = 1

  for i in range(len(array)):
    if freq_dict[array[i]] == 1:
      return i
  return -1
```

```python
print(find_first_unique_index([1, 2, 3]))
print(find_first_unique_index([1, 2, 3, 1]))
print(find_first_unique_index([1, 2, 3, 1, 2]))
print(find_first_unique_index([1, 2, 3, 1, 2, 3]))
```
