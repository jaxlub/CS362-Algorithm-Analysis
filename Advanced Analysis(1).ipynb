# Binary search

This function implements our binary search algorithm. Assume that array is sorted.

```python
def binary_search(target, array):
  left = 0
  right = len(array) - 1

  while left <= right:

    mid = (left+right)//2
    if array[mid] > target:
      right = mid - 1
    elif array[mid] < target:
      left = mid + 1
    else:
      return True
  return False
```

```python
ls = [1, 3, 5, 7]
for value in range(9):
  print(value, binary_search(value, ls))
```
