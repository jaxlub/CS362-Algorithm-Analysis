# Interval selection

This function implements our greedy heuristic for selecting as many non-overlapping intervals as possible.

```python
def greedy_most_intervals(intervals):
  selected = list()
  for interval in sorted(intervals, key =lambda t: t[1]):
    if len(selected) == 0 or interval[0] >= selected[-1][1]:
      selected.append(interval)
  return selected

```

```python
print(greedy_most_intervals([(0,6), (1,4), (3,5), (3,8), (4,7), (5,9), (6,10), (8,11)]))
```

# Change making

This function implements our greedy heuristic for making change.

```python
def greedy_change(n, coins):
  change = list()
  for coin in sorted(coins, reverse = True):
    while coin <= n:
      change.append(coin)
      n = n - coin
  return change
```

```python
print(greedy_change(6, {1, 5, 10, 25}))
print(greedy_change(12, {1, 5, 10, 25}))
print(greedy_change(31, {1, 5, 10, 25}))
```

```python
print(greedy_change(6, {25, 10, 6, 1}))
print(greedy_change(12, {25, 10, 6, 1}))
print(greedy_change(31, {25, 10, 6, 1}))
```
