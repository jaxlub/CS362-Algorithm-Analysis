# Problem 1

Given an array of length $n$, determine whether its values are ordered from smallest to largest.

A) In a few sentences, give an algorithm idea for this task.

Loop through the array from the beggening and compare the previous value to the next and if the previous is bigger then return false. If end of array is reached return true.

B) Analyze your algorithm.

O(n) where n is the number of elements in the array. As the array grows so too does the time. Worst case scenario the entire array is ordered smallest to largest so indexes must be checked. Best case scenario the first value is bigger then the second and fails without further looping and is therefore $\Omega$(1) or constant.


C) Implement your algorithm.

```python
def check_sorted(array):
  if array != []:
    for i in range (1, len(array)):
      if array[i-1] > array[i]:
        return False
  return True
```

```python
# These arrays are sorted.
print(check_sorted([]))
print(check_sorted([1]))
print(check_sorted([1, 2]))
print(check_sorted([1, 2, 3]))
print(check_sorted([1, 2, 3, 3]))
```

```python
# These arrays are not sorted.
print(check_sorted([2, 1]))
print(check_sorted([1, 3, 2]))
print(check_sorted([1, 2, 4, 3]))
```

# Problem 2

Given an array of length $n$, determine whether it contains any duplicate values.

A) In a few sentences, give an algorithm idea for this task.

loop through the array and compare each value to all values *after* it in the array for duplicates. If at any point a duplicate is found, return true. Once all combos are touched and no true has been returned then return false.

B) Analyze your algorithm.

O(n^2) because you must loop through the array and for each iteration you must loop through the rest of the array so the nested loops make it n^2 where n is the size of the array. Now in the best case scenario a duplicate is found right away which eliminates any need for further computation and thus has a Î©(1) or constant.

C) Implement your algorithm.

```python
def check_duplicates(array):
  if array != []:
    for i in range(len(array)-1):
      for y in range(i+1, len(array)):
        if array[i] == array[y]:
          return True
  return False
```

```python
# These arrays contain duplicates.
print(check_duplicates([1, 1, 2, 3]))
print(check_duplicates([1, 2, 1, 3]))
print(check_duplicates([1, 2, 3, 1]))
print(check_duplicates([2, 1, 1, 3]))
print(check_duplicates([2, 1, 3, 1]))
print(check_duplicates([2, 3, 1, 1]))
```

```python
# These arrays do not contain duplicates.
print(check_duplicates([]))
print(check_duplicates([1]))
print(check_duplicates([1, 2]))
print(check_duplicates([1, 2, 3]))
```
